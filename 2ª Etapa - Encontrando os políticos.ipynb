# Código utilizado para análise de dados da reportagem: [Prefeito e 5 vereadores eleitos em 2024 estão na lista suja do trabalho escravo](https://g1.globo.com/politica/noticia/2025/01/28/prefeito-e-5-vereadores-eleitos-em-2024-estao-na-lista-suja-do-trabalho-escravo.ghtml).
## Neste segundo código, os dados de autuados por trabalho escravo (tabela feita no primeiro código) sçao cruzados com os [dados eleitorais de 2024 disponibilizados pelo TSE](https://dadosabertos.tse.jus.br/dataset/candidatos-2024). 

```python
# Importa as bibliotecas necessárias para a análise
import pandas as pd
import os
```

```python
# Me mostre tudo
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
```

```python
# Acessando dados de sócios autuados
socios = pd.read_csv("data/socios-fichasuja.csv", delimiter=',', encoding='utf-8', dtype=str)
```

```python
socios.head()
```

```python
# Padronizando para maiúsculas
socios = socios.astype(str).applymap(str.upper)
```

```python
# Acessando base de dados de candidatos nas eleições 2024
eleitos = pd.read_csv("data/consulta_cand_2024_BRASIL.csv", delimiter =';', encoding='latin-1')
```

```python
eleitos.head()
```

```python
# Consultando as situações de candidatura dos políticos
eleitos.DS_SIT_TOT_TURNO.unique()
```

```python
# Selecionando apenas os eleitos e suplentes
eleitos = eleitos[eleitos['DS_SIT_TOT_TURNO'].isin([
    'ELEITO POR MÉDIA',
    'ELEITO POR QP',
    'SUPLENTE', 
    'ELEITO'])]
```

```python
# Cruzando tabela de sócios autuados com tabela políticos eleitos
df_final = pd.merge(socios, eleitos, left_on='socio', right_on='NM_CANDIDATO', how='left')
```

```python
df_final.head()
```

```python
# Salvar o resultado em um arquivo csv
df_final.to_csv('data/politicos_autuados.csv', decimal=";", index=False, encoding='utf-8')
```
